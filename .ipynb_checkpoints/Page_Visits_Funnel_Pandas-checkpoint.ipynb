```python
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
#1 Inspect the DataFrames using print and head: 
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))

#2 Combine visits and cart using a left merge.
visits_and_cart = pd.merge(visits, cart, how='left')
print(visits_and_cart)

#3 How long is your merged DataFrame?
num_visits_and_cart = len(visits_and_cart)
print(num_visits_and_cart)

#4 How many of the timestamps are null for the column cart_time? Not adding to cart.
num_dont_add_to_cart = len(visits_and_cart[visits_and_cart.cart_time.isnull()])
print(num_dont_add_to_cart)

#5 What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?
pc_dont_add_to_cart = (float(num_dont_add_to_cart)/float(num_visits_and_cart)) *100
print(str(pc_dont_add_to_cart) + "% of visitors don't add to cart")

#6 Repeat the left merge for cart and checkout and count null values. What percentage of users put items in their cart, but did not proceed to checkout?
cart_and_checkout = pd.merge(cart, checkout, how='left')
print(cart_and_checkout)
num_cart_and_checkout = len(cart_and_checkout)
num_dont_checkout = len(cart_and_checkout[cart_and_checkout.checkout_time.isnull()])
pc_dont_checkout = round((float(num_dont_checkout)/float(num_cart_and_checkout))*100, 2)
print(str(pc_dont_checkout) + "% of visitors who add to cart don't checkout")

#7 Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data. Examine the result using print and head.
all_data = visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
all_data.drop_duplicates(subset='user_id', inplace=True)
print(all_data.head(10))

#8 What percentage of users proceeded to checkout, but did not purchase a t-shirt?
num_of_checkout = len(checkout)
num_of_purchase = len(purchase)
num_checkout_no_purchase = num_of_checkout - num_of_purchase
pc_checkout_no_purchase = round((num_checkout_no_purchase/num_of_purchase)*100,2)
print(str(pc_checkout_no_purchase) + "% of visitors who proceeded to checkout do not purchase")

#9 Which step of the funnel is weakest (i.e., has the highest percentage of users not completing it)?How might Cool T-Shirts Inc. change their website to fix this problem?
pc_funnel_drop = [pc_dont_add_to_cart, pc_dont_checkout, pc_checkout_no_purchase]
print("Highest funnel percentage drop is " + str(max(pc_funnel_drop)) +"%")

#10 Using the giant merged DataFrame all_data that you created, letâ€™s calculate the average time from initial visit to final purchase. Add a column that is the difference between purchase_time and visit_time. 
all_data['visit_to_purchase_time'] = all_data.purchase_time - all_data.visit_time

#11 Examine the results by printing the new column to the screen.
print(all_data['visit_to_purchase_time'])

# 12 Calculate the average time to purchase by applying the .mean() function to your new column.
average_purchase_time = all_data['visit_to_purchase_time'].mean()
print("Average time to purchase the item is " + str(average_purchase_time))
```
